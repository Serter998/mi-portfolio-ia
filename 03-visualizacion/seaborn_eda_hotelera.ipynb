## Carga de Datos

Ejecuta esta celda para cargar el dataset:


````python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
n = 1200

# Generar datos de encuestas
meses = np.random.choice(['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic'], n)
tipo_cliente = np.random.choice(['Familiar', 'Pareja', 'Negocio', 'Solo'], n, p=[0.35, 0.30, 0.20, 0.15])
canal = np.random.choice(['Web', 'Booking', 'Agencia', 'Directo'], n, p=[0.40, 0.30, 0.15, 0.15])
noches = np.random.choice([1,2,3,4,5,6,7], n, p=[0.05,0.15,0.25,0.25,0.15,0.10,0.05])

# Satisfacción base que varía por tipo de cliente
base = {'Familiar': 7.5, 'Pareja': 8.0, 'Negocio': 7.0, 'Solo': 7.2}
satisfaccion = np.array([base[t] + np.random.normal(0, 1.2) for t in tipo_cliente])
satisfaccion = np.clip(satisfaccion, 1, 10).round(1)

# Precio correlacionado con noches y tipo
precio_base = 80 + noches * 45 + np.random.normal(0, 30, n)
precio = np.clip(precio_base, 50, 600).round(0)

# NPS derivado de satisfacción
nps = np.where(satisfaccion >= 9, 'Promotor',
      np.where(satisfaccion >= 7, 'Pasivo', 'Detractor'))

# Repetidor correlacionado con satisfacción
prob_repetir = (satisfaccion - 5) / 10
repetidor = np.random.binomial(1, np.clip(prob_repetir, 0.05, 0.95))

edad = np.random.normal(42, 12, n).astype(int)
edad = np.clip(edad, 18, 80)

df = pd.DataFrame({
    'mes': meses,
    'tipo_cliente': tipo_cliente,
    'canal': canal,
    'noches': noches,
    'satisfaccion': satisfaccion,
    'precio': precio,
    'nps': nps,
    'repetidor': repetidor,
    'edad': edad
})

# Orden de meses
orden_meses = ['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']
df['mes'] = pd.Categorical(df['mes'], categories=orden_meses, ordered=True)

print("Dataset Hotel Costa Brava cargado correctamente")
print(f"Forma: {df.shape}")
print(df.head())
````
